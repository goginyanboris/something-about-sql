# Задание 1

Необходимо вывести названия треков, которые были приобретены компаниями, которые обслуживаются сотрудниками, нанятыми после 1 сентября 2002 года. 
    

```python
%%sql
Select c.Customerid, c.company, e.HireDate 
from customers c
Join employees e
ON c.SupportRepId = e.EmployeeId AND e.HireDate > "2002-09-01" AND c.company != "None"
```

```python
%%sql
Select i.invoiceid, i.customerid, a.company
from invoices i
    Join(
    Select c.Customerid, c.company
    from customers c
    Join employees e
    ON c.SupportRepId = e.EmployeeId AND e.HireDate > "2002-09-01" AND c.company != "None"
) a
ON i.Customerid = a.Customerid
```

```python
%%sql
Select *
from invoice_items ii
Join(
    Select i.invoiceid, i.customerid, a.company
    from invoices i
        Join(
        Select c.Customerid, c.company
        from customers c
        Join employees e
        ON c.SupportRepId = e.EmployeeId AND e.HireDate > "2002-09-01" AND c.company != "None"
    ) a
    ON i.Customerid = a.Customerid
)b
ON b.invoiceid = ii.invoiceid
```

```python
%%sql
Select t.name, c.company
from tracks t
join(
    Select *
    from invoice_items ii
    Join(
        Select i.invoiceid, i.customerid, a.company
        from invoices i
            Join(
            Select c.Customerid, c.company
            from customers c
            Join employees e
            ON c.SupportRepId = e.EmployeeId AND e.HireDate > "2002-09-01" AND c.company != "None"
        ) a
        ON i.Customerid = a.Customerid
    )b
    ON b.invoiceid = ii.invoiceid
) c
ON t.TrackId = c.TrackId
```

## Задание 2
Вывести общую стоимость и продолжительность каждого плейлиста.

```python
%%sql
Select SUM(t.unitprice) as "общая стоимость", SUM(t.milliseconds) as "общее время", count(t.trackid) as "кол-во песен в плейлисте", a.playlistid, a.name as "плейлист"
From Tracks t
Join(
    Select pt.playlistid, pt.TrackId, p.Name
    from playlist_track pt
    join playlists p
    ON pt.PlaylistId = p.PlaylistId
) a
ON t.trackid = a.trackid
Group by a.PlaylistId
```

## Задание 5
Для каждой компании вывести общее количестов песен, купленных по жанру рок

```python
%%sql 
SELECT d.company, f.song_count  --d.customerid, f.cust
FROM customers d
left join(
        SELECT g.name, c.customerid as cust, c.company as co, count(1) as song_count
        FROM genres g
        JOIN(
            SELECT t.TrackId, t.GenreId as GenreId, b.InvoiceId, b.CustomerId, b.Company
            FROM tracks t
            JOIN(
                SELECT ii.InvoiceId, ii.TrackId as TrackId, a.CustomerId, a.Company
                FROM Invoice_items ii
                JOIN(
                     SELECT i.InvoiceId, i.CustomerId, c.company
                     FROM invoices i
                     JOIN customers c
                     ON i.CustomerId = c.CustomerId AND c.company != "None"
                    ) a
                ON ii.InvoiceId = a.InvoiceId
                ) b
            ON t.TrackId = b.TrackId
            ) c
        ON g.GenreId = c.GenreId
        WHERE g.name = "Rock"
        GROUP BY c.company
    ) f
ON f.Cust = d.CustomerId
GROUP BY d.company
Having d.company != "None"
```
