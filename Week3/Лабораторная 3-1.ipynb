# Задание 1

Вывести 10 самых больших по размеру треков жанра ROCK и формата MPEG (исоп

```python
%%sql
SELECT t.name, t.genreid, t.mediatypeid, t.bytes -- таблица с жанром "рок" и форматом "мпег аудио файл(тк нужна музыка)", 
FROM tracks t
JOIN genres g 
ON g.GenreId = t.GenreId AND g.name = "Rock"
JOIN media_types m
ON t.mediatypeid = m.mediatypeid AND m.name = "MPEG audio file"
```

```python
%%sql -- (ИТОГ)
SELECT name, bytes
FROM (SELECT t.name as name, t.genreid, t.mediatypeid, t.bytes as bytes
        FROM tracks t
        JOIN genres g 
        ON g.GenreId = t.GenreId AND g.name = "Rock"
        JOIN media_types m
        ON t.mediatypeid = m.mediatypeid AND m.name = "MPEG audio file"
    )
ORDER BY bytes DESC -- по убыванию первые 10
LIMIT 10
```

```python
%%sql
SELECT *
FROM invoice_items ii
JOIN(SELECT *
     FROM invoices i
     JOIN customers c
     ON i.CustomerId = c.CustomerId AND c.company = "Microsoft Corporation"
    ) k
ON ii.Invoiceid = k.InvoiceId
```

```python
%%sql
SELECT *
FROM tracks t
JOIN(
    SELECT ii.TrackId
    FROM invoice_items ii
    JOIN(SELECT *
         FROM invoices i
         JOIN customers c
         ON i.CustomerId = c.CustomerId AND c.company = "Microsoft Corporation" -- получены все треки, приобретенные микрософт жанра рок
        ) k
    ON ii.Invoiceid = k.InvoiceId
    )p
ON t.TrackId = p.TrackId
JOIN genres g 
ON g.GenreId = t.GenreId AND g.name = "Rock"
```

```python
%%sql -- (ИТОГ)
SELECT t.name as "Song name", a.title as "Album", ar.name as "Group"
FROM tracks t
JOIN(
    SELECT ii.TrackId
    FROM invoice_items ii
    JOIN(SELECT *
         FROM invoices i
         JOIN customers c
         ON i.CustomerId = c.CustomerId AND c.company = "Microsoft Corporation" -- получены все треки, приобретенные микрософт жанра рок
        ) k
    ON ii.Invoiceid = k.InvoiceId
    )p
ON t.TrackId = p.TrackId
JOIN genres g 
ON g.GenreId = t.GenreId AND g.name = "Rock"
JOIN albums a
ON a.AlbumId = t.AlbumId 
JOIN artists ar
ON a.artistid = ar.artistid
```

## Задание 3
Для каждого набора (жанр, тип медиа) вывести среднюю цену по стоимости трека и общее количество, причем вывести только те наборы, для которых все треку стоят больше 1,5$.


```python
%%sql
SELECT AVG(UnitPrice) as "AVG", count(t.name) as "obshee kolvo trekov", g.name as genre, m.name as tip_media
FROM tracks t
JOIN genres g, media_types m
ON g.genreid = t.genreid AND m.MediaTypeId = t.MediaTypeId
GROUP BY g.name, m.name
Having MIN(t.unitprice) > 1.5
```

## Задание 4
Вывести компании, сделавшие максимальное и минимальное число заказов.

```python
%%sql
    SELECT count(c.company) as ccc, c.company
    FROM invoices i
    JOIN customers c
    ON i.CustomerId = c.CustomerId
    WHERE c.company <> "None"
    GROUP BY c.company
```

```python
%%sql   -- (ИТОГ)
SELECT cm.company
FROM(
    SELECT count(c.company) as ccc, c.company
    FROM invoices i
    JOIN customers c
    ON i.CustomerId = c.CustomerId
    WHERE c.company <> "None"
    GROUP BY c.company
    )cm
GROUP BY cm.company
HAVING MAX(cm.ccc)

UNION

SELECT cm.company
FROM(
    SELECT count(c.company) as ccc, c.company
    FROM invoices i
    JOIN customers c
    ON i.CustomerId = c.CustomerId
    WHERE c.company <> "None"
    GROUP BY c.company
    )cm
GROUP BY cm.company
HAVING MIN(cm.ccc)
```

```python
%%sql
SELECT ii.InvoiceId, ii.TrackId as TrackId, a.CustomerId, a.Company
FROM Invoice_items ii
JOIN(
     SELECT i.InvoiceId, i.CustomerId, c.company
     FROM invoices i
     JOIN customers c
     ON i.CustomerId = c.CustomerId AND c.company != "None"
    ) a
ON ii.InvoiceId = a.InvoiceId
```

```python
%%sql
SELECT t.TrackId, t.GenreId, b.InvoiceId, b.CustomerId, b.Company
FROM tracks t
JOIN(
    SELECT ii.InvoiceId, ii.TrackId as TrackId, a.CustomerId, a.Company
    FROM Invoice_items ii
    JOIN(
         SELECT i.InvoiceId, i.CustomerId, c.company
         FROM invoices i
         JOIN customers c
         ON i.CustomerId = c.CustomerId AND c.company != "None"
        ) a
    ON ii.InvoiceId = a.InvoiceId
    ) b
ON t.TrackId = b.TrackId
```

```python
%%sql
SELECT *  --, count(g.name)
FROM genres g
JOIN(
    SELECT t.TrackId, t.GenreId as GenreId, b.InvoiceId, b.CustomerId, b.Company
    FROM tracks t
    JOIN(
        SELECT ii.InvoiceId, ii.TrackId as TrackId, a.CustomerId, a.Company
        FROM Invoice_items ii
        JOIN(
             SELECT i.InvoiceId, i.CustomerId, c.company
             FROM invoices i
             JOIN customers c
             ON i.CustomerId = c.CustomerId AND c.company != "None"
            ) a
        ON ii.InvoiceId = a.InvoiceId
        ) b
    ON t.TrackId = b.TrackId
    ) c
ON g.GenreId = c.GenreId
```

```python
%%sql
SELECT g.name, c.customerid as cust, c.company as co, count(1) as song_count
FROM genres g
JOIN(
    SELECT t.TrackId, t.GenreId as GenreId, b.InvoiceId, b.CustomerId, b.Company
    FROM tracks t
    JOIN(
        SELECT ii.InvoiceId, ii.TrackId as TrackId, a.CustomerId, a.Company
        FROM Invoice_items ii
        JOIN(
             SELECT i.InvoiceId, i.CustomerId, c.company
             FROM invoices i
             JOIN customers c
             ON i.CustomerId = c.CustomerId AND c.company != "None"
            ) a
        ON ii.InvoiceId = a.InvoiceId
        ) b
    ON t.TrackId = b.TrackId
    ) c
ON g.GenreId = c.GenreId
WHERE g.name = "Pop"
GROUP BY c.company
```

```python
%%sql //(ИТОГ)
SELECT d.company, f.song_count  --d.customerid, f.cust
FROM customers d
left join(
        SELECT g.name, c.customerid as cust, c.company as co, count(1) as song_count
        FROM genres g
        JOIN(
            SELECT t.TrackId, t.GenreId as GenreId, b.InvoiceId, b.CustomerId, b.Company
            FROM tracks t
            JOIN(
                SELECT ii.InvoiceId, ii.TrackId as TrackId, a.CustomerId, a.Company
                FROM Invoice_items ii
                JOIN(
                     SELECT i.InvoiceId, i.CustomerId, c.company
                     FROM invoices i
                     JOIN customers c
                     ON i.CustomerId = c.CustomerId AND c.company != "None"
                    ) a
                ON ii.InvoiceId = a.InvoiceId
                ) b
            ON t.TrackId = b.TrackId
            ) c
        ON g.GenreId = c.GenreId
        WHERE g.name = "Pop"
        GROUP BY c.company
    ) f
ON f.Cust = d.CustomerId
GROUP BY d.company
Having d.company != "None"
```

## Задание 6
Вывести средний размер альбома в байтах.

```python
%%sql
SELECT a.title as "название альбома", count(a.title) as "кол-во альбомов", AVG(t.bytes) as "средний размер"
FROM tracks t
JOIN albums a
ON t.albumid = a.albumid
Group By a.title
```
